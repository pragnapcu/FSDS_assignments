```python
base = int(input("Enter base value of a triangle:"))
height = int(input("Enter height value of a triangle:"))
area = (base * height)/2
print("The area of a triangle is:", area)
```

```python
4.	Write a Python program to swap two variables?
```

```python
a = 10
b = 20
print("Entered value of a: ", a)
print("Entered value of b: ", b)
c = b
b = a
a = c
print("Swapped value of a: ", a)
print("Swapped value of b: ", b)
```
