```python
1.	Write a Python Program to Find the Factorial of a Number?
```

```python
num =  int(input("Enter a number: "))
fac = 1

for i in range(1, num+1):
    fac = fac * i

print(f'Factorial of {num} is {fac}')
```

```python
3.	Write a Python Program to Print the Fibonacci sequence?
```

```python
num2 =  int(input("Enter a number: "))
a = 0
b = 1
count = 2
fib = [a, b]
while(count<num2):
    c= a + b
    fib.append(c)
    count = count + 1
    a = b
    b = c

print(fib)
```

```python
4.	Write a Python Program to Check Armstrong Number?
```

```python
num3 = input("Enter a number: ")
n = list(num3)
mult = len(num3)
armstrong = 0
for i in range(0, len(n)):
#     print(n[i])
    armstrong = armstrong + ((int(n[i])) ** mult)

if int(num3) == armstrong:
    print(f'{num3} is an armstrong number')
else:
    print(f'{num3} is not an armstrong number')
```

```python
5.	Write a Python Program to Find Armstrong Number in an Interval?
```

```python
armstrng = 0

for i in range(100, 10000):
    int_2_str = str(i)          # int to str bcz int is not iterable
    str_2_lst = list(int_2_str) # str to list to separate each digit of a number
    len_i = len(str_2_lst)      #length of list to use for multiplication while calculating artmstrong

    for j in range(0, len_i):
        armstrng = armstrng + int(str_2_lst[j]) ** len_i #str to int; power

    if armstrng == i:
        print(f'{i} is an armstrong number')
    armstrng = 0
```

```python
6.	Write a Python Program to Find the Sum of Natural Numbers?
```

```python
num4 = int(input("Enter a number: "))
sum = 0

for i in range(1, num4+1):
    sum = sum + i

print(f'The sum of {num4} natural numbers is {sum}')
```
